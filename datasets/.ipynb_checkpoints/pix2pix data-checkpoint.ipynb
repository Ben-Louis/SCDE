```python
class Pix2PixData(Dataset):
    def __init__(self, obj='shoes', mode='train', image_size=128):
        
        self.image_size = image_size
        
        # dir
        assert obj in ['shoes', 'handbags']
        root = os.path.join(os.environ['HOME'], 'dataset', 'pix2pix', 'datasets', 'edges2'+obj)
        train_dir = os.path.join(root, 'train')
        valid_dir = os.path.join(root, 'val')
    
        # files
        train_files = [os.path.join(train_dir, fname) for fname in os.listdir(train_dir)]
        valid_files = [os.path.join(valid_dir, fname) for fname in os.listdir(valid_dir)]
        if mode == 'train':
            self.files = train_files
        else:
            self.files = valid_files                
        
        # transforms
        self.toTensor = transforms.Compose([
            transforms.Resize((image_size, image_size*2),Image.BILINEAR),
            transforms.ToTensor()])
        self.norm = transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
        
    def __len__(self):
        return len(self.files)
    
    def __getitem__(self, index):
        img = Image.open(self.files[index])
        img = self.toTensor(img)
        
        skt, pho = img[:,:,:self.image_size], img[:,:,self.image_size:]
        skt = torch.mean(skt, dim=0, keepdim=True)
        skt = 1-(skt>0.8).type(torch.FloatTensor)
        pho = self.norm(pho)
        
        return pho, skt
```

```python
from torch.autograd import Variable
```

```python
skts = Variable(skts)
```

```python
wid, sigma = 9, 0.2
```

```python
wg = torch.from_numpy(cv2.getGaussianKernel(wid, sigma))
```

```python
wg = wg.mm(torch.t(wg)).type(torch.FloatTensor)
```

```python
w = wg.unsqueeze(0).unsqueeze(0)
```

```python
skts_blur = torch.nn.functional.conv2d(skts, Variable(w), stride=1, padding=wid//2)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(121), plt.imshow(skts[2][0].data.numpy(), 'gray')
plt.subplot(122), plt.imshow(skts_blur[2][0].data.numpy(), 'gray')
```
